$\textbf{Exercício 1}:$ Indique se as afirmações abaixo são verdadeiras ou falsas, e justifique brevemente.

(a) Se um modelo tem um erro de teste baixo, o valor do erro de treino não traz nenhuma informação
adicional sobre como potencialmente melhorar o modelo.

Falso. Ainda que o erro de teste seja satisfatório, as comparações abaixo podem ser feitas:

- Erro de teste e treino são iguais: o modelo performa bem tanto no conjunto amostral em que foi treinado quanto em observações desconhecidas, sinalizando que é um modelo com capacidade de generalizar o universo de possibilidades sob análise.
- Erro de teste menor que o erro de treino: as métricas de distribuição da amostra de treino e de testes devem ser analisadas para que o analista possa entender o motivo dessa diferença. Se a amostra de testes tiver uma distribuição mais `comportada` em termos de média e desvio-padrão do que a amostra de treino, vale repensar o método de seleção das amostras para que elas tenham características mais semelhantes e, portanto, mais generalizáveis.
- Erro de teste maior que o erro de treino: ainda se o erro de teste seja considerado `satisfatório`, o modelo pode estar apresentado um $\textbf{overfit}$. Essa informação pode justificar uma análise de outros problemas comuns em modelos estatísticos, como correlação entre os erros (especialmente se o conjunto de dados for uma série temporal ou espacial). Outra fonte comum de *overfitting* é a alta dimensionalidade de modelos, que pode ser reduzida através de análises de colinearidade.

(b) Numa regressão linear, não é possível testar uma hipótese $H_0: \beta_j = 0$ usando o teste $t$ sem fazer alguma hipótese sobre a distribuição dos erros.

Verdadeiro. A adoção de uma regressão linear implica a hipótese de que os erros apresentam distribuição normal. Por sua vez, $t_j = \hat{\beta}_j / \mathrm{SE}(\hat{\beta}_j)$ de forma que o valor $t$ representa quantos erros padrão o coeficiente está do ponto nulo. Se a distribuição dos erros não for normal (e.g., uma `power law`), a estatística $t$ pode resultar em uma sensação de confiança no modelo injustificada.

Se o modelo tiver baixa dimensionalidade (i.e., $p < 3$), é possível plotar e observar a distribuição dos erros e ajustar os dados, a depender do formato da distribuição. Por exemplo, se os *targets* (Y) apresentarem uma distribuição exponencial, é possível realizar uma transformação usando funções concavas (e.g., $log(Y)$ ou $sqrt(Y)$) ou convexas (e.g., $Y^r$).

(c) Um banco de varejo está interessado em reduzir transações fraudulentas e desenvolveu dois modelos para isso; o primeiro tem acurácia (isto é, um menos a perda 0-1) 98.5% e o outro 99%.
Suponha que o banco se preocupe mais em barrar todas as transações fraudulentas do que em aceitar todas as transações legítimas. Nesse caso, é necessariamente preferível usar o modelo de
maior acurácia.

Falso. O banco tem maior tolerância com erros do Tipo I ou falsos positivos (modelo aponta como fraudulenta uma transação legítima) do que com erros do Tipo II ou falsos negativos (modelo aponta como legítima uma transação fraudulenta). Nesse cenário, essas situações precisam ser analisadas de forma independente, através de métricas como *precision* e *recall*, obtidas através de uma *confusion matrix*, que basicamente indica onde o modelo está errando.

- Acurácia: $\frac{Previsões corretas}{Total de previsões}$ - útil como indicativo geral de performance do modelo. Pode-se dizer que uma acurácia alta é condição necessária, mas não suficiente, para a adoção de um modelo.
- Precision: $\frac{Positivos verdadeiros}{Total de positivos}$
- Recall: $\frac{Positivos verdadeiros}{}$
